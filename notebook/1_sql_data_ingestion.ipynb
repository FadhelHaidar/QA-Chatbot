```python
import kagglehub

path = kagglehub.dataset_download("kartik2112/fraud-detection")
path
```

```python
import pandas as pd
import glob
```

```python
data_files = glob.glob(path + "/*.csv")
data_files
```

```python
df = pd.read_csv(data_files[0], index_col=0)
df
```

```python
fraud_df = df[df['is_fraud'] == 1]
fraud_df
```

```python
no_fraud_df = df[df['is_fraud'] == 0]
no_fraud_df
```

```python
concatenated_df = pd.concat([fraud_df[:50], no_fraud_df[:50]])
concatenated_df
```

```python
engine = "postgresql://myuser:mypassword@localhost:5432/mydb"
concatenated_df.to_sql("fraud_data", con=engine, index=False, if_exists="replace")
```
