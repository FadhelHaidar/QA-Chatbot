```python
from langchain_community.document_loaders import DirectoryLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
loader = DirectoryLoader('../data/txt', glob='**/*.txt')
docs = loader.load()
len(docs)
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size = 1024,
    chunk_overlap = 256
)

splits = text_splitter.split_documents(docs)
splits
```

# Ingestion and Embedding

```python
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_qdrant import QdrantVectorStore
from qdrant_client import QdrantClient
from qdrant_client.http.models import Distance, VectorParams
```

```python
embedding = HuggingFaceEmbeddings(model_name="Qwen/Qwen3-Embedding-0.6B")
```

```python
url = "localhost:6334"
```

```python
qdrant = QdrantVectorStore.from_documents(
    documents=splits,
    embedding=embedding,
    url=url,
    prefer_grpc=True,
    collection_name="my_documents",
)
```

# Indexing

```python
qdrant_store = QdrantVectorStore(
    client=QdrantClient(url=url, prefer_grpc=True),
    collection_name="my_documents",
    embedding=embedding
)
```

```python
qdrant_store.similarity_search("how fraud is committed worldwide?", k=5)
```
